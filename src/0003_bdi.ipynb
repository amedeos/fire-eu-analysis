# BDI - Extract MFN_BMK.M.020.922.0.EUR.210 Data

This notebook extracts data from the BDI Excel report, specifically the column MFN_BMK.M.020.922.0.EUR.210 along with the date column DATA_OSS.

The output CSV will contain the date and value columns, with backward fill (bfill) applied to handle missing values.


```python
import pandas as pd
import os
import warnings
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Suppress openpyxl style warnings (innocuous warnings about missing default styles)
warnings.filterwarnings('ignore', category=UserWarning, module='openpyxl')
```

```python
# Change to project root directory
# Find the project root by looking for the 'data' directory
current_dir = os.getcwd()
while not os.path.exists(os.path.join(current_dir, 'data')):
    parent_dir = os.path.dirname(current_dir)
    if parent_dir == current_dir:
        # Reached filesystem root without finding 'data' directory
        raise FileNotFoundError("Could not find project root directory (looking for 'data' folder)")
    current_dir = parent_dir

os.chdir(current_dir)
print(f"Working directory: {os.getcwd()}")

# Define input and output files
input_file = 'data/bdi/BMK0100/REPORT_06-12-2025_11-27_1765016876665.xlsx'
output_file = 'data/bdi/BMK0100/MFN_BMK.M.020.922.0.EUR.210.csv'

# Check that the input file exists
if not os.path.exists(input_file):
    raise FileNotFoundError(f"File not found: {input_file}")

print(f"Input file: {input_file}")
print(f"Output file: {output_file}")
```

```python
# Read the Excel file without header to analyze structure
print("Reading Excel file (first 5 rows to analyze structure)...")
df_raw = pd.read_excel(input_file, header=None, nrows=5)

print(f"First 5 rows (raw):")
print(df_raw.to_string())
print(f"\nAnalyzing structure...")

# Find the row with DATA_OSS (should be row 2, column 0)
# Find the column with MFN_BMK.M.020.922.0.EUR.210 (should be row 1)
date_col_idx = None
value_col_idx = None

# Look for DATA_OSS in row 2 (index 2)
if len(df_raw) > 2:
    row2 = df_raw.iloc[2]
    for idx, val in enumerate(row2):
        if pd.notna(val) and 'DATA_OSS' in str(val):
            date_col_idx = 0  # DATA_OSS is always in column 0
            print(f"Found DATA_OSS in row 2, column {date_col_idx}")
            break

# Look for MFN_BMK.M.020.922.0.EUR.210 in row 1 (index 1)
if len(df_raw) > 1:
    row1 = df_raw.iloc[1]
    for idx, val in enumerate(row1):
        if pd.notna(val) and 'MFN_BMK.M.020.922.0.EUR.210' in str(val):
            value_col_idx = idx
            print(f"Found MFN_BMK.M.020.922.0.EUR.210 in row 1, column {value_col_idx}")
            break

if date_col_idx is None or value_col_idx is None:
    raise ValueError(f"Could not find required columns. Date col: {date_col_idx}, Value col: {value_col_idx}")

# Now read the full file, skipping the first 2 rows (rows 0 and 1) and using row 2 as header
# But we'll manually set column names
print(f"\nReading full file (skipping first 2 rows, data starts at row 3)...")
df = pd.read_excel(input_file, header=None, skiprows=2)

# Set column names manually
df.columns = [f'col_{i}' for i in range(len(df.columns))]

# Extract only the date column (index 0) and value column
date_col_name = df.columns[date_col_idx]
value_col_name = df.columns[value_col_idx]

print(f"Using columns: {date_col_name} (date), {value_col_name} (value)")
print(f"\nData shape: {df.shape}")
print(f"\nFirst few rows:")
print(df[[date_col_name, value_col_name]].head())
```

```python
# Extract the required columns and rename them
df_extracted = df[[date_col_name, value_col_name]].copy()
df_extracted.columns = ['DATA_OSS', 'MFN_BMK.M.020.922.0.EUR.210']

date_col = 'DATA_OSS'
value_col = 'MFN_BMK.M.020.922.0.EUR.210'

print(f"Extracted columns:")
print(f"  Date column: {date_col}")
print(f"  Value column: {value_col}")

print(f"\nExtracted data shape: {df_extracted.shape}")
print(f"\nFirst few rows of extracted data:")
print(df_extracted.head())
print(f"\nData info:")
print(df_extracted.info())
print(f"\nMissing values:")
print(df_extracted.isnull().sum())
```

```python
# Remove the header row if it exists (contains "DATA_OSS" or "VALORE")
df_extracted = df_extracted[
    ~df_extracted[date_col].astype(str).str.contains('DATA_OSS', na=False)
].copy()

# Clean the date column - it may contain strings like "[2025-10-31 00:00:00] 2025-ott"
# Extract just the date part
def clean_date(date_val):
    if pd.isna(date_val):
        return None
    date_str = str(date_val)
    # Try to extract date from format like "[2025-10-31 00:00:00] 2025-ott"
    if '[' in date_str and ']' in date_str:
        # Extract the part between [ and ]
        start = date_str.find('[') + 1
        end = date_str.find(']')
        date_str = date_str[start:end].split()[0]  # Take first part before space
    return date_str

# Apply date cleaning
df_extracted[date_col] = df_extracted[date_col].apply(clean_date)

# Convert date column to datetime with explicit format to avoid warnings
# Format is YYYY-MM-DD (e.g., 2025-10-31)
df_extracted[date_col] = pd.to_datetime(df_extracted[date_col], format='%Y-%m-%d', errors='coerce')

# Remove rows where date is invalid
df_extracted = df_extracted[df_extracted[date_col].notna()].copy()

# Sort by date to ensure proper order
df_extracted = df_extracted.sort_values(by=date_col).reset_index(drop=True)

print(f"Date range: {df_extracted[date_col].min()} to {df_extracted[date_col].max()}")
print(f"Total rows: {len(df_extracted)}")
print(f"\nMissing values before bfill:")
print(df_extracted.isnull().sum())
```

```python
# Convert value column to numeric (removing any non-numeric rows like header)
df_extracted[value_col] = pd.to_numeric(df_extracted[value_col], errors='coerce')

# Apply backward fill (bfill) to fill missing values
# This fills missing values with the next available value
df_extracted[value_col] = df_extracted[value_col].bfill()

print(f"Missing values after bfill:")
print(df_extracted.isnull().sum())
print(f"\nFinal data preview:")
print(df_extracted.head(10))
print(f"\n...")
print(df_extracted.tail(10))
```

```python
# Create output directory if it doesn't exist
os.makedirs(os.path.dirname(output_file), exist_ok=True)

# Save to CSV
df_extracted.to_csv(output_file, index=False)
print(f"File saved successfully: {output_file}")
print(f"Dimensions: {len(df_extracted)} rows, {len(df_extracted.columns)} columns")
print(f"\nFinal data summary:")
print(df_extracted.describe())
```

```python
# Create a plot of the data
plt.figure(figsize=(14, 7))
plt.plot(df_extracted[date_col], df_extracted[value_col], linewidth=1.5, color='#2E86AB')
plt.title('BDI - MFN_BMK.M.020.922.0.EUR.210 (BTP Decennale Benchmark) - Andamento Storico', 
          fontsize=14, fontweight='bold')
plt.xlabel('Data', fontsize=12)
plt.ylabel('Rendimento Lordo (%)', fontsize=12)
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()

# Format x-axis dates
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator(2))  # Show every 2 years
plt.xticks(rotation=45)

# Save plot
plot_file = os.path.join(os.path.dirname(output_file), 'MFN_BMK.M.020.922.0.EUR.210_plot.png')
plt.savefig(plot_file, dpi=150, bbox_inches='tight')
print(f"Plot saved to: {plot_file}")

plt.show()
```

# BDI - Extract MFN_BMK.D.020.922.0.EUR.210 Data (Daily)

This section processes daily data from multiple Excel files in the BMK0200 directory.


```python
# Function to extract data from a single Excel file
def extract_bdi_data(excel_file, value_col_name_pattern):
    """
    Extract DATA_OSS and value column from a BDI Excel file.
    
    Parameters:
    -----------
    excel_file : str
        Path to the Excel file
    value_col_name_pattern : str
        Pattern to search for in the value column (e.g., 'MFN_BMK.D.020.922.0.EUR.210')
    
    Returns:
    --------
    pd.DataFrame with columns DATA_OSS and the value column
    """
    # Read first 5 rows to analyze structure
    df_raw = pd.read_excel(excel_file, header=None, nrows=5)
    
    # Find the column indices
    date_col_idx = 0  # DATA_OSS is always in column 0
    value_col_idx = None
    
    # Look for the value column in row 1 (index 1)
    if len(df_raw) > 1:
        row1 = df_raw.iloc[1]
        for idx, val in enumerate(row1):
            if pd.notna(val) and value_col_name_pattern in str(val):
                value_col_idx = idx
                break
    
    if value_col_idx is None:
        raise ValueError(f"Could not find column '{value_col_name_pattern}' in file {excel_file}")
    
    # Read the full file, skipping the first 2 rows
    df = pd.read_excel(excel_file, header=None, skiprows=2)
    df.columns = [f'col_{i}' for i in range(len(df.columns))]
    
    # Extract only the date and value columns
    date_col_name = df.columns[date_col_idx]
    value_col_name = df.columns[value_col_idx]
    
    df_extracted = df[[date_col_name, value_col_name]].copy()
    df_extracted.columns = ['DATA_OSS', value_col_name_pattern]
    
    # Remove header row if it exists
    df_extracted = df_extracted[
        ~df_extracted['DATA_OSS'].astype(str).str.contains('DATA_OSS', na=False)
    ].copy()
    
    # Clean dates
    def clean_date(date_val):
        if pd.isna(date_val):
            return None
        date_str = str(date_val)
        if '[' in date_str and ']' in date_str:
            start = date_str.find('[') + 1
            end = date_str.find(']')
            date_str = date_str[start:end].split()[0]
        return date_str
    
    df_extracted['DATA_OSS'] = df_extracted['DATA_OSS'].apply(clean_date)
    df_extracted['DATA_OSS'] = pd.to_datetime(df_extracted['DATA_OSS'], format='%Y-%m-%d', errors='coerce')
    df_extracted = df_extracted[df_extracted['DATA_OSS'].notna()].copy()
    
    # Convert value to numeric
    df_extracted[value_col_name_pattern] = pd.to_numeric(df_extracted[value_col_name_pattern], errors='coerce')
    
    return df_extracted

print("Function defined successfully")
```

```python
# Process all daily files in BMK0200 directory
input_dir = 'data/bdi/BMK0200'
output_file_daily = 'data/bdi/BMK0200/MFN_BMK.D.020.922.0.EUR.210.csv'
value_col_pattern = 'MFN_BMK.D.020.922.0.EUR.210'

# Get all Excel files in the directory
excel_files = [f for f in os.listdir(input_dir) if f.endswith('.xlsx')]
excel_files.sort()  # Sort to process in order

print(f"Found {len(excel_files)} Excel files in {input_dir}:")
for f in excel_files:
    print(f"  - {f}")

if len(excel_files) == 0:
    raise FileNotFoundError(f"No Excel files found in {input_dir}")

# Process each file
all_data = []
for excel_file in excel_files:
    file_path = os.path.join(input_dir, excel_file)
    print(f"\nProcessing {excel_file}...")
    try:
        df_file = extract_bdi_data(file_path, value_col_pattern)
        df_file['source_file'] = excel_file  # Track source file for debugging
        all_data.append(df_file)
        print(f"  Extracted {len(df_file)} rows, date range: {df_file['DATA_OSS'].min()} to {df_file['DATA_OSS'].max()}")
    except Exception as e:
        print(f"  Error processing {excel_file}: {e}")
        raise

print(f"\nTotal files processed: {len(all_data)}")
```

```python
# Combine all dataframes
df_combined = pd.concat(all_data, ignore_index=True)

# Remove source_file column (was only for debugging)
df_combined = df_combined[['DATA_OSS', value_col_pattern]].copy()

print(f"Combined data shape: {df_combined.shape}")
print(f"Date range: {df_combined['DATA_OSS'].min()} to {df_combined['DATA_OSS'].max()}")
print(f"\nMissing values before deduplication:")
print(df_combined.isnull().sum())

# Check for duplicate dates and verify consistency
print(f"\nChecking for duplicate dates...")
duplicate_dates = df_combined[df_combined.duplicated(subset=['DATA_OSS'], keep=False)].copy()

if len(duplicate_dates) > 0:
    print(f"Found {len(duplicate_dates)} rows with duplicate dates")
    print("\nAnalyzing duplicates for consistency...")
    
    # Group by date and check consistency
    date_groups = df_combined.groupby('DATA_OSS')
    inconsistencies = []
    
    for date, group in date_groups:
        if len(group) > 1:
            values = group[value_col_pattern].dropna().unique()
            # Check if all non-null values are the same
            if len(values) > 1:
                inconsistencies.append({
                    'date': date,
                    'values': values.tolist(),
                    'count': len(group)
                })
            # If one is null and others are not, that's also inconsistent (unless all are null)
            null_count = group[value_col_pattern].isna().sum()
            non_null_count = group[value_col_pattern].notna().sum()
            if null_count > 0 and non_null_count > 0:
                # This is acceptable: same date, some have value, some don't
                # We'll take the non-null value
                pass
    
    if inconsistencies:
        print(f"\nWARNING: Found {len(inconsistencies)} dates with inconsistent values:")
        for inc in inconsistencies[:10]:  # Show first 10
            print(f"  Date {inc['date']}: values {inc['values']}")
        if len(inconsistencies) > 10:
            print(f"  ... and {len(inconsistencies) - 10} more")
    else:
        print("All duplicate dates have consistent values (same value or missing in all)")
    
    # Deduplicate: keep first non-null value, or first if all are null
    df_combined = df_combined.sort_values(by=['DATA_OSS', value_col_pattern], 
                                          na_position='last').drop_duplicates(subset=['DATA_OSS'], keep='first')
    print(f"\nAfter deduplication: {len(df_combined)} rows")
else:
    print("No duplicate dates found")

# Sort by date
df_combined = df_combined.sort_values(by='DATA_OSS').reset_index(drop=True)

print(f"\nFinal data shape: {len(df_combined)} rows")
print(f"Missing values before bfill: {df_combined[value_col_pattern].isnull().sum()}")
```

```python
# Apply backward fill (bfill) to fill missing values
df_combined[value_col_pattern] = df_combined[value_col_pattern].bfill()

print(f"Missing values after bfill: {df_combined[value_col_pattern].isnull().sum()}")
print(f"\nFinal data preview:")
print(df_combined.head(10))
print(f"\n...")
print(df_combined.tail(10))
```

```python
# Create output directory if it doesn't exist
os.makedirs(os.path.dirname(output_file_daily), exist_ok=True)

# Save to CSV
df_combined.to_csv(output_file_daily, index=False)
print(f"File saved successfully: {output_file_daily}")
print(f"Dimensions: {len(df_combined)} rows, {len(df_combined.columns)} columns")
print(f"\nFinal data summary:")
print(df_combined.describe())
```

```python
# Create a plot of the daily data
plt.figure(figsize=(16, 8))
plt.plot(df_combined['DATA_OSS'], df_combined[value_col_pattern], linewidth=0.8, color='#A23B72', alpha=0.7)
plt.title('BDI - MFN_BMK.D.020.922.0.EUR.210 (BTP Decennale Benchmark - Dati Giornalieri) - Andamento Storico', 
          fontsize=14, fontweight='bold')
plt.xlabel('Data', fontsize=12)
plt.ylabel('Rendimento Lordo (%)', fontsize=12)
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()

# Format x-axis dates
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator(2))  # Show every 2 years
plt.xticks(rotation=45)

# Save plot
plot_file_daily = os.path.join(os.path.dirname(output_file_daily), 'MFN_BMK.D.020.922.0.EUR.210_plot.png')
plt.savefig(plot_file_daily, dpi=150, bbox_inches='tight')
print(f"Plot saved to: {plot_file_daily}")

plt.show()
```
